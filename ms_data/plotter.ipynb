```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['figure.dpi'] = 120
```

```python
df1 = pd.read_csv('dat1_ox5.csv', header=0, delimiter='\t')
df2 = pd.read_csv('dat2_ox10.csv', header=0, delimiter='\t')
df3 = pd.read_csv('dat3_ox20.csv', header=0, delimiter='\t')
df4 = pd.read_csv('dat4_ox40.csv', header=0, delimiter='\t')
df5 = pd.read_csv('dat5_ox80.csv', header=0, delimiter='\t')
```

```python
t1 = df1['t'].to_numpy()
p1 = (df1['p'] - df1['p'][len(df1['p'])-1]).to_numpy()

t2 = df2['t'].to_numpy()
p2 = (df2['p'] - df2['p'][len(df2['p'])-1]).to_numpy()

t3 = df3['t'].to_numpy()
p3 = (df3['p'] - df3['p'][len(df3['p'])-1]).to_numpy()

t4 = df4['t'].to_numpy()
p4 = (df4['p'] - df4['p'][len(df4['p'])-1]).to_numpy()

t5 = df5['t'].to_numpy()
p5 = (df5['p'] - df5['p'][len(df5['p'])-1]).to_numpy()
```

```python
plt.plot(t1, p1, label='5 min oxidation')
plt.plot(t2, p2, label='10 min oxidation')
plt.plot(t3, p3, label='20 min oxidation')
plt.plot(t4, p4, label='40 min oxidation')
plt.plot(t5, p5, label='80 min oxidation')

plt.legend()

plt.xlabel('time [s]')
plt.ylabel('pressure [Torr]')

plt.xlim(0.0, 6500)
plt.ylim(0.0, 2.1e-8)
# plt.savefig('offset_removed.png')
```

```python
# df1 start coord: (809.344, 1.2208e-8) index: 403
# df2 start coord: (1001.281, 1.2583e-8) index: 499
# df3 start coord: (951.141 1.2558e-8) index: 475
# df4 start coord: (1169.157, 1.4692e-8) index: 583
# df5 start coord: (1024.594, 1.093e-8) index: 511

p1 = 100 * p1/p1[403]
p2 = 100 * p2/p2[499]
p3 = 100 * p3/p3[475]
p4 = 100 * p4/p4[583]
p5 = 100 * p5/p5[511]
```

```python
plt.plot(t1, p1, label='5 min oxidation')
plt.plot(t2, p2, label='10 min oxidation')
plt.plot(t3, p3, label='20 min oxidation')
plt.plot(t4, p4, label='40 min oxidation')
plt.plot(t5, p5, label='80 min oxidation')

plt.legend()

plt.xlabel('time [s]')
plt.ylabel('pressure [Torr]')

plt.xlim(0.0, 6000)
plt.ylim(0.0, 160)
plt.savefig('normalized.png')
```
